```python
"""
"*" or "-" or "+" 연산자가 등장할때 좌/우 분할을 하고

각각 계산 결과를 리턴함
"""
from typing import List

def diffWaysToCompute(input: str) -> List[int]:
  def compute(left, right, op):
    results = []

    for l in left:
      for r in right:
        results.append(eval(str(l) + op + str(r)))

    return results

  if input.isdigit():
    return [int(input)]

  results = []

  for index, value in enumerate(input):
    if value in "-+*":
      left = diffWaysToCompute(input[:index])
      right = diffWaysToCompute(input[index + 1:])

      results.extend(compute(left, right, value))

  return results

# result = [0, 2]
# 2 - 1 - 1
"""
(2 - 1) - 1 = 0

2 - (1 - 1) = 2
"""
# input = "2-1-1"


# result = [-34, -14, -10, -10, 10]
# 2 x 3 - 4 x 5
"""
2 x (3 - (4 x 5)) = -34

(2 x 3) - (4 x 5) = -14

(2 x (3 - 4)) x 5 = -10

2 x ((3 - 4) x 5) = -10

((2 x 3) - 4) x 5 = 10
"""
input = "2*3-4*5"

print(diffWaysToCompute(input))
```
