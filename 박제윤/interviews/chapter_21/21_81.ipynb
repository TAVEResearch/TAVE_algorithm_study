```python
# 풀이 1: Visit all of them
# That is, let's solve it by designating one space as the starting point from the beginning and visiting all the remaining gas stations. 
# O(n^2)
from typing import List
import time

def canComplete(gas: List[int], cost: List[int]) -> int:
  for start in range(len(gas)):
    fuel = 0

    for i in range(start, len(gas) + start):
      index = i % len(gas)
      # print("idx:", index)

      can_travel = True
      if gas[index] + fuel < cost[index]:
        can_travel = False
        break
      else:
        fuel += gas[index] - cost[index]

    if can_travel:
      return start
  return -1

# result = 3
gas = [1, 2, 3, 4, 5]
cost = [3, 4, 5, 1, 2]

current_time = time.time()
print(canComplete(gas, cost))
end_time = time.time() - current_time
print(end_time)
```

```python
# 풀이 2: Visit once
# If the total amount of oil is greater than the total cost, 
# there must be a starting point to visit all of them.
from typing import List

def canComplete2(gas: List[int], cost: List[int]) -> int:
  # whether it can tour all the gas stations 
  if sum(gas) < sum(cost):
    return -1

  start, fuel = 0, 0

  for i in range(len(gas)):
    # whether it's not appropriate as a starting point
    if gas[i] + fuel < cost[i]:
      start = i + 1
      fuel = 0
    else:
      fuel += gas[i] - cost[i]
    
  return start

# result = 3
gas = [1, 2, 3, 4, 5]
cost = [3, 4, 5, 1, 2]

current_time = time.time()
print(canComplete2(gas, cost))
end_time = time.time() - current_time
print(end_time)
print(0.0012767314910888672 / end_time)
```
