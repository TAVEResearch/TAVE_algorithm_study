# 7. 배열

- 자료구조는 크게 두가지 방식으로 나뉨

  (1) 메모리 공간 기반의 연속 (Contiguous) 방식

  (2) 포인터 기반의 연결 (Link) 방식

- 여기서 배열은 연속 방식의 가장 기본이 되는 자료형

- 배열은 스택이나 큐 같은 **Abstract Data Type (ADT)**를 구현할때도 사용됨

- 오늘날 32비트 이상의 시스템에서 int는 4바이트를 사용함

- 메모리에 대한 접근은 1바이트씩 하므로 주소는 1바이트당 1씩 증가함

- 배열은 어느 위치에서나 $O(1)$ 에 조회가 가능하다는 장점이 있음

  e.g.) 위의 그림의 배열에서 4번째 값에 접근하고 싶으면 int 배열이므로 각 4바이트 $(4 - 1) \times 4 = 12$ 가 되고, 0x00에서 시작해 12만큼 증가한 16진수는 0x0c가 됨

- 파이썬에선 배열 (a.k.a 리스트)은 동적 배열 자료형임

- 즉, 미리 크기를 지정하지 않고, 자동으로 리사이징 함

- 동적 배열은 초기값을 작게 잡아 배열을 생성하고 데이터가 추가되면 사이즈를 늘리고 (e.g. Doubling, 즉 2배씩 늘림) 모두 복사하는식으로 동작함
