## 풀이1) 재귀 구조 부루트 포스 

Capture.PNG

```python
"""
Disclaimer: 타임아웃!😥
"""
def climbStairs(n: int) -> int:
  # exception
  if n == 1:
    return 1
  if n == 2:
    return 2
  
  return climbStairs(n - 1) + climbStairs(n - 2)

# result = 3
n = 3
print(climbStairs(n))
```

## 풀이 2) 메모이제이션 (Top - Down)

```python
from collections import defaultdict

def climbStairs_sol2(n: int) -> int:
  dp = defaultdict(int)

  if n <= 2:
    return n
  
  # dp가 비어있지 않으면 반환
  if dp[n]:
    return dp[n]

  dp[n] = climbStairs(n - 1) + climbStairs(n - 2)
  

  return dp[n]

result = 3
# n = 10
print(climbStairs_sol2(n))
```
