### 2) W != H (직사각형)

- 규칙성을 모르겠음...

- 자세히 보니까 일차함수

- 좌표평면 놓고 해볼까?

- 일차함수 구해놓고 $1 \times 1$이니까 x 좌표 하나씩 넣으면서 y값의 정수값 다 더하고 대칭이니까 마지막에 2만 곱하면 될듯!!!

```python
# date: 2022.03.17
# author: jeiyoon
def solution(W: int, H: int) -> int:
  # exception
  if W == H: # square
    return W * (H - 1)
  if W == 1 or H == 1:
    return 0

  answer = 0

  # inclination
  inclination = -1 * float(H / W)
    
  for w in range(1, W): 
    print(w)
    print(int(inclination * w + H))
    answer += int(inclination * w + H) # y = ax + b   

  return answer * 2

# result = 80
W = 8
H = 12
print(solution(W, H))
```
