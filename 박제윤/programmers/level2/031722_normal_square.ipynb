```python
# date: 2022.03.17
# author: jeiyoon
def solution(W: int, H: int) -> int:
  # exception
  if W == H: # square
    return W * (H - 1)
  if W == 1 or H == 1:
    return 0

  answer = 0

  # inclination
  inclination = -1 * float(H / W)
    
  for w in range(1, W): 
    print(w)
    print(int(inclination * w + H))
    answer += int(inclination * w + H) # y = ax + b   

  return answer * 2

# result = 80
W = 8
H = 12
print(solution(W, H))
```
