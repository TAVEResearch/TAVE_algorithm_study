# 78. 주식을 사고 팔기 가장 좋은 시점

### 1. 그리디 알고리즘

* 내리기 전에 팔고, 오르기 전에 산다.

* 오르기 전 값을 사고, 내리기 전 값을 판다

* 두점을 이은 선의 기울기가 상향일때 상황만 계산하여 result 값에 더한다.

```python
from typing import List


class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        result = 0
        # 값이 오르는 경우 매 번 그리디 계산
        for i in range(len(prices) - 1):
            if prices[i + 1] > prices[i]:
                result += prices[i + 1] - prices[i]
        return result
    
sol = (Solution().maxProfit([7,1,5,3,6,4]))
print(sol)
```

### 2. 파이썬 다운 방식

```python
from typing import List


class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        # 0보다 크면 무조건 합산
        return sum(max(prices[i + 1] - prices[i], 0) for i in range(len(prices) - 1))

sol = (Solution().maxProfit([7,1,5,3,6,4]))
print(sol) 
```
