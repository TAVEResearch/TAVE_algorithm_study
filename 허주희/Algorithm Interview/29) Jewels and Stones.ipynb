## 1) 해시 테이블을 이용한 풀이

```python
class Solution:
    def numJewelsInStones(self, J, S) :
        freqs = {}
        count = 0

        # 돌(S)의 빈도 수 계산
        for char in S:
            if char not in freqs:
                freqs[char] = 1
            else:
                freqs[char] += 1
        print(freqs)
        
        # 보석(J)의 빈도 수 합산
        for char in J:
            if char in freqs:
                count += freqs[char]

        return count

J = "aA"
S = "aAAbbbb"
print(Solution().numJewelsInStones(J, S))
```

> 돌(S)의 빈도 수 계산

```python
class Solution:
    def numJewelsInStones(self, J, S) :
        freqs = {}
        count = 0

        # 돌(S)의 빈도 수 계산
        for char in S:
            if char not in freqs:
                ## freqs 리스트 안에 char가 없으면 새로 추가
                freqs[char] = 1
            else:
                freqs[char] += 1  
                ## freqs 리스트 안에 char가 있으면 개수 +1

        return freqs

J = "aA"
S = "aAAbbbb"
print(Solution().numJewelsInStones(J, S))
```

## 2) defaultdict를 이용한 비교 생략

```python
import collections


class Solution:
    def numJewelsInStones(self, J: str, S: str) -> int:
        freqs = collections.defaultdict(int)
        count = 0

        # 비교 없이 돌(S) 빈도 수 계산
        for char in S:
            freqs[char] += 1

        # 비교 없이 보석(J) 빈도 수 합산
        for char in J:
            count += freqs[char]

        return count
```

defaultdict(int) : 기본값 0으로 존재하지 않는 키에 대해 리턴

## 3) counter를 이용한 비교 생략

```python
import collections


class Solution:
    def numJewelsInStones(self, J: str, S: str) -> int:
        freqs = collections.Counter(S)  # 돌(S) 빈도 수 계산
        count = 0
        print(freqs)
        # 비교 없이 보석(J) 빈도 수 합산
        for char in J:
            count += freqs[char]

        return count

J = "aA"
S = "aAAbbbb"
print(Solution().numJewelsInStones(J, S))
```

counter : 자동으로 개수 계산
