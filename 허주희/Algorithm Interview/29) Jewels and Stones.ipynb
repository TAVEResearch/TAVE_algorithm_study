```python
class Solution:
    def numJewelsInStones(self, J, S) :
        freqs = {}
        count = 0

        # (S)   
        for char in S:
            if char not in freqs:
                freqs[char] = 1
            else:
                freqs[char] += 1
        print(freqs)
        
        # (J)   
        for char in J:
            if char in freqs:
                count += freqs[char]

        return count

J = "aA"
S = "aAAbbbb"
print(Solution().numJewelsInStones(J, S))
```

## 2) defaultdict   

```python
import collections


class Solution:
    def numJewelsInStones(self, J: str, S: str) -> int:
        freqs = collections.defaultdict(int)
        count = 0

        #   (S)   
        for char in S:
            freqs[char] += 1

        #   (J)   
        for char in J:
            count += freqs[char]

        return count
```

## 3) counter
