```python
s="{{2},{2,1},{2,1,3},{2,1,3,4}}"


def solution(s):
    answer = []
    new_s = s[2:-2]
    ## "2},{2,1},{2,1,3},{2,1,3,4"
    new_s = new_s.split('},{')
    ##['2', '2,1', '2,1,3', '2,1,3,4']
    while True:
        if len(new_s) == 1:
            answer.append(int(new_s[0]))
            break
        else:
            first(new_s)
            answer.append(int(one))
    return answer


def first(new_s):
    global one

    for i in range(len(new_s) - 1):
        num = new_s[i]
        l = len(num)
        if l == 1:
            one = num
            del new_s[i]
        else:
            continue
        ##    , 
        ## ['2,1', '2,1,3', '2,1,3,4']

        for i in range(len(new_s)):
            word = new_s[i]
            if one in word:
                new_s[i] = word.strip(one)
                new_s[i] = word.strip(',' + one)
                new_s[i] = word.strip(one + ',')
        ##        
        ## ['1', '1,3', '1,3,4']
        return new_s


print(solution(s))
```
